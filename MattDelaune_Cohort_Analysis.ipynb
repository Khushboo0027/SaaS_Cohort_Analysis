# Methodology

This project employs a comprehensive cohort analysis to uncover actionable insights into customer engagement, retention, and revenue generation within a SaaS business context. The analysis is structured around two key dimensions: **first purchase dates** and **hypothesized product usage groups**, allowing for a granular examination of customer behavior over time.

**Approach:**
- **Customer Segmentation:** Customers are categorized into distinct groups based on their purchasing behavior and product usage patterns. This segmentation enables a more focused analysis of trends and behaviors within each group, facilitating the development of tailored strategies to enhance engagement and retention.
- **Cohort Analysis:** By tracking the performance of these customer segments across various time periods, the analysis provides a clear understanding of how different strategies impact long-term customer value. This method is particularly effective for identifying the factors that drive retention and profitability, offering insights into both immediate outcomes and evolving trends.

**Key Metrics:**
The analysis focuses on the following metrics to provide a comprehensive view of customer dynamics:
- **Customer Retention Rate (CRR):** Tracks the percentage of customers within each cohort who continue to make purchases over time. This metric is crucial for assessing customer loyalty and the success of retention strategies.
- **Net Revenue Retention (NRR):** Measures the percentage of recurring revenue retained from existing customers, accounting for expansions (upsells), contractions (downsells), and churn. NRR provides a holistic view of revenue stability and growth within each cohort.
- **Customer Lifetime Revenue (CLR):** Calculates the total revenue generated by a customer throughout their relationship with the business. This metric helps in identifying high-value segments that contribute significantly to overall revenue.
- **Customer Lifetime Value (CLV):** Estimates the total profit a business can expect from a customer over their entire relationship. CLV is a critical metric for guiding strategic decisions on customer acquisition, retention investments, and resource allocation.

**Relevance:**
This methodology is designed to answer the core business question by revealing both immediate and long-term trends in customer behavior. It provides a robust framework for understanding the drivers of customer value, allowing for strategic decisions that align with business objectives and drive profitability.

<a id="renewal-cycle-estimation"></a>
### **Renewal Cycle Estimation**

Understanding that different SaaS products may have varying renewal cycles based on their nature, contract terms, and usage patterns, we will estimate the renewal cycles for each group. This estimation involves calculating the most common interval in months between consecutive purchases, after filtering out anomalies (such as purchases made within 1-2 days or over a year apart, which are considered irregular). The intervals are then converted from days to months.

Accurately estimating the renewal cycle enables more meaningful segmentation and comparison across different product groups. By comprehending typical renewal patterns, we can more precisely calculate retention rates, enhancing the granularity of net revenue retention, customer lifetime revenue, and customer lifetime value across various segments. This approach facilitates identifying which product groups excel in customer retention and revenue generation.

```python
# Function to calculate renewal intervals for each product
def calculate_product_renewal_intervals(df):
    df = df.sort_values(by=['Customer ID', 'Order Date'])
    
    df['Previous Order Date'] = df.groupby(['Customer ID', 'Product'])['Order Date'].shift(1)
    df['Days Between Purchases'] = (df['Order Date'] - df['Previous Order Date']).dt.days
    
    # Filter out anomalies (e.g., purchases made within 1-2 days might not be renewals but additional orders)
    df = df[(df['Days Between Purchases'] > 2) & (df['Days Between Purchases'] < 365)]
    
    # Convert days to months
    df['Months Between Purchases'] = (df['Days Between Purchases'] / 30).round()
    df = df[df['Months Between Purchases'] > 1]
    
    # Determine common renewal intervals in months for each product
    def get_mode(series):
        mode_values = series.mode()
        if len(mode_values) > 0:
            return mode_values.iloc[0]
        return None
    
    renewal_intervals = df.groupby('Product')['Months Between Purchases'].agg(get_mode).reset_index()
    return renewal_intervals

# Calculate and print refined renewal intervals for each product
product_renewal_intervals = calculate_product_renewal_intervals(df)
print(product_renewal_intervals)

# Compare the renewal intervals within each group
for group_name, products in zip(['Core Services', 'Add-ons or Plugins', 'Specialized Tools'], 
                                [core_services, addons_plugins, specialized_tools]):
    group_intervals = product_renewal_intervals[product_renewal_intervals['Product'].isin(products)]
    print(f"\n{group_name} Renewal Intervals:\n{group_intervals}")
```

The data suggests that a 2-month renewal interval is a strong general assumption for most products across different categories. Core services, add-ons, and specialized tools generally follow this 2-month cycle, indicating a common usage pattern among customers. While real SaaS renewal cycles may vary, this fictional dataset appears to be designed with uniform renewal cycles.

We will proceed using this 2-month interval as a baseline assumption for calculating our metrics. This interval will be recalculated for every subset of data to ensure accuracy and relevance.

### **Cohort Definition: First Purchase Month**

While sign-up date and subscription-based information are ideal for defining cohorts and calculating metrics, we will use the first purchase date as a proxy, assuming that clients typically make their first purchase shortly after signing up.

To determine first purchase dates, we will convert the Order Date into monthly periods, group by Customer ID, and identify the earliest transaction for each customer. This approach will provide insights into customer behavior following their initial purchase, enabling us to analyze revenue, customer engagement, and retention.

```python
# Create OrderMonth for analysis
df['OrderMonth'] = df['Order Date'].dt.to_period('M')

# Create FirstPurchaseMonth
df['FirstPurchaseMonth'] = df.groupby('Customer ID')['Order Date'].transform('min').dt.to_period('M')

# Define Cohort
df['Cohort'] = df['FirstPurchaseMonth']
```

<a id="customer-retention-rate-crr"></a>
## **Customer Retention Rate (CRR)**

<a id="crr-by-first-purchase-cohort"></a>
### **CRR By First Purchase Cohort**

```python
# Group by Cohort and OrderMonth to get the number of unique customers
cohort_data = df.groupby(['Cohort', 'OrderMonth']).agg(n_customers=('Customer ID', 'nunique')).reset_index()

# Calculate the period index (difference in months) adjusted for 2-month intervals
cohort_data['PeriodIndex'] = ((cohort_data['OrderMonth'].dt.year - cohort_data['Cohort'].dt.year) * 12 + 
                              (cohort_data['OrderMonth'].dt.month - cohort_data['Cohort'].dt.month)) // 2

# Pivot the data for retention matrix
cohort_pivot = cohort_data.pivot_table(index='Cohort', columns='PeriodIndex', values='n_customers')

# Calculate the retention rate
cohort_size = cohort_pivot.iloc[:, 0]
retention_matrix = cohort_pivot.divide(cohort_size, axis=0)

# Plot the retention matrix
plt.figure(figsize=(20, 12))
sns.heatmap(data=retention_matrix, annot=True, fmt='.0%', cmap=custom_palette, annot_kws={"size": 6}, linewidths=0.5)
plt.title('Customer Retention Rate by First Purchase Cohort', fontsize=18)
plt.xlabel('Period Index (2-Month Intervals)', fontsize=14)
plt.ylabel('Cohort', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.show()
```

When we look at the earlier cohorts, such as January 2020, we see strong retention rates at the beginning. Some of these cohorts maintain retention rates above 100%, thanks to repeat or upsell purchases. But as time goes on, we notice a gradual decline in these rates, such as in February and March 2020, which shows that while initial engagement is solid, customers tend to drop off over time.

The cohorts from April to May 2020 tell a slightly different story. They show more inconsistency, with some experiencing rapid declines in retention early on. This could be due to differences in customer satisfaction or how useful they find the product.

One of the key insights here is that there are significant drop-offs in retention around the 6 to 12-month mark (or the 3rd to 6th period). This period seems to be critical for customer churn. It underscores how crucial it is to keep customers engaged and deliver ongoing value during these months to improve retention.

After this initial drop, some cohorts do stabilize in their retention rates, while others continue to decline. This variation suggests that customer satisfaction levels differ, and we might need more targeted strategies to keep different groups of customers.

Using 85% to 90% retention annually as a [SaaS industry benchmark](https://blog.hubspot.com/service/customer-retention) to compare our data to, we see that, despite a wide range of retention percentages shown, several cohorts fall below our ideal mark.

<a id="clr-vs-clv-cost-efficiency"></a>
### **CLR vs. CLV Cost Efficiency**

Here I will compare the CLR and CLV trends, along with their Revenue to Cost Ratio and Profit to Cost Ratio, to determine if high revenue translates into high profitability for each usage group.

```python
# Function to plot average lifetime revenue, profit, and cost efficiency by product category
def plot_category_lifetime_metrics(product_category, category_name):
    category_data = df[df['Product'].isin(product_category)]
    
    # Recreate OrderMonth and Cohort for the filtered dataset
    category_data['OrderMonth'] = category_data['Order Date'].dt.to_period('M')
    category_data['FirstPurchaseMonth'] = category_data.groupby('Customer ID')['Order Date'].transform('min').dt.to_period('M')
    category_data['Cohort'] = category_data['FirstPurchaseMonth']
    
    # Calculate cumulative revenue, profit, and cost per customer
    category_data['CustomerCumulativeRevenue'] = category_data.groupby('Customer ID')['Sales'].cumsum()
    category_data['CustomerCumulativeProfit'] = category_data.groupby('Customer ID')['Profit'].cumsum()
    category_data['CustomerCumulativeCost'] = category_data['CustomerCumulativeRevenue'] - category_data['CustomerCumulativeProfit']  # Cost = Sales - Profit
    
    # Group by Customer and Cohort to get the lifetime revenue, profit, and cost
    clv_data = category_data.groupby(['Customer ID', 'Cohort']).agg(
        lifetime_revenue=('CustomerCumulativeRevenue', 'max'),
        lifetime_profit=('CustomerCumulativeProfit', 'max'),
        lifetime_cost=('CustomerCumulativeCost', 'max')
    ).reset_index()

    # Group by Cohort to get the average lifetime revenue, profit, and cost
    cohort_clv = clv_data.groupby('Cohort').agg(
        avg_lifetime_revenue=('lifetime_revenue', 'mean'),
        avg_lifetime_profit=('lifetime_profit', 'mean'),
        avg_lifetime_cost=('lifetime_cost', 'mean')
    ).reset_index()

    # Calculate the Revenue to Cost Ratio (RCR) and Profit to Cost Ratio (PCR)
    cohort_clv['RCR'] = cohort_clv['avg_lifetime_revenue'] / cohort_clv['avg_lifetime_cost']
    cohort_clv['PCR'] = cohort_clv['avg_lifetime_profit'] / cohort_clv['avg_lifetime_cost']
    
    # Calculate the number of 2-month periods since the cohort started
    cohort_start = cohort_clv['Cohort'].min()
    cohort_clv['PeriodIndex'] = cohort_clv['Cohort'].apply(lambda x: ((x.start_time.year - cohort_start.start_time.year) * 12 + x.start_time.month - cohort_start.start_time.month) // 2)

    # Plot the average lifetime revenue and profit on the primary y-axis
    fig, ax1 = plt.subplots(figsize=(12, 8))

    ax1.set_title(f'Average Customer Lifetime Revenue, Value, and Cost Efficiency by Cohort for {category_name}')
    ax1.set_xlabel('2-Month Periods Since Cohort Start')
    ax1.set_ylabel('Average Value')
    
    sns.lineplot(data=cohort_clv, x='PeriodIndex', y='avg_lifetime_revenue', label='Average Lifetime Revenue', color=custom_palette[1], ax=ax1, ci = None)
    sns.lineplot(data=cohort_clv, x='PeriodIndex', y='avg_lifetime_profit', label='Average Lifetime Value (Profit)', color=custom_palette[5], ax=ax1, ci = None)

    # Create a secondary y-axis for RCR and PCR
    ax2 = ax1.twinx()
    ax2.set_ylabel('Cost Efficiency Ratios')
    
    sns.lineplot(data=cohort_clv, x='PeriodIndex', y='RCR', label='Revenue to Cost Ratio (RCR)', color=custom_palette[1], ax=ax2, linestyle='--', ci = None)
    sns.lineplot(data=cohort_clv, x='PeriodIndex', y='PCR', label='Profit to Cost Ratio (PCR)', color=custom_palette[5], ax=ax2, linestyle='-.', ci = None)
    
    fig.tight_layout()
    fig.legend().remove()
    plt.show()

    # Calculate and print KPIs
    avg_rcr = cohort_clv['RCR'].mean()
    avg_pcr = cohort_clv['PCR'].mean()
    print(f"KPIs for {category_name}:")
    print(f"Average Revenue to Cost Ratio (RCR): {avg_rcr:.2f}")
    print(f"Average Profit to Cost Ratio (PCR): {avg_pcr:.2f}")
    print("\n")    
    
# Plot lifetime revenue, profit, and cost efficiency for each product category
plot_category_lifetime_metrics(core_services, 'Core Services')
plot_category_lifetime_metrics(addons_plugins, 'Add-ons or Plugins')
plot_category_lifetime_metrics(specialized_tools, 'Specialized Tools')

```

For core services, the average RCR of 1.13 indicates that for every dollar spent, the service generates $1.13 in revenue. This is relatively healthy, indicating a small margin over cost. However, the average PCR of 0.15 suggests that profit margins are quite thin. In comparison to SaaS industry standards, a typical RCR is around 1.25 to 1.50, while a healthy PCR would ideally be above 0.25 [(HubSpot Blog)](https://blog.hubspot.com/service/saas-metrics). This suggests room for improvement in profitability for core services.

The add-ons or plugins category has a better performance with an RCR of 1.25, aligning more closely with industry standards. The PCR of 0.28 indicates a healthier profit margin compared to core services. These metrics suggest that the add-ons or plugins are performing well in terms of cost efficiency and profitability, likely due to their higher perceived value or lower associated costs.

Specialized tools have an RCR of 1.13, similar to core services, but their PCR of 0.21 is better, suggesting moderate profitability. Although this is below the ideal industry threshold, it indicates that specialized tools are more efficient in converting costs to profits compared to core services but are less efficient than add-ons or plugins.

The RCR values across all product categories indicate that they are generating revenue slightly above costs, which is good. However, for core services and specialized tools, efforts should be made to either increase the revenue per unit cost or decrease the cost itself to push the RCR closer to the industry standard of 1.25 to 1.50.
The PCR values suggest that add-ons or plugins have the healthiest profit margins. For core services and specialized tools, the focus should be on strategies to increase profitability. This can be achieved by optimizing pricing models, reducing costs, or increasing customer lifetime value through better service offerings and customer retention strategies.
The company should consider investing more in add-ons or plugins, as these categories show the highest cost efficiency and profitability. Efforts to improve the profitability of core services and specialized tools should also be prioritized to ensure a balanced and profitable portfolio.
